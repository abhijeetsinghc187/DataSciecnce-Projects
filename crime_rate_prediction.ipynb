Creaate Sample Dataset


```python
import pandas as pd

data = {
    "Date": ["2020-01-01","2020-02-01","2020-03-01","2021-01-01","2021-02-01"],
    "Primary Type": ["Theft","Assault","Robbery","Theft","Burglary"],
    "Location Description": ["Street","Home","Shop","Street","Home"]
}

df = pd.DataFrame(data)
df.to_csv("crime_data.csv", index=False)

df.head()
```

Preprocessing


```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Hour'] = 12 #dummy hour

#  Convert Categorical to numbers

df['Primary Type'] = df['Primary Type'].astype('category').cat.codes
df['Location Description'] = df['Location Description'].astype('category').cat.codes

df.head()
```

Train Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x = df[['Year','Month','Hour','Primary Type','Location Description']]
y = df ['Primary Type']
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2)

model = RandomForestRegressor()

model.fit(x_train, y_train)

print("MOdel Trained Successfully!")
```

Evaluate Model

```python
from sklearn.metrics import r2_score

preds = model.predict(x_test)
print("R2 Score:", r2_score(y_test,preds))
```

Save Model

```python
import pickle
with open("model.pk1", "wb") as f:
  pickle.dump(model,f)
  print("Model Saved")
```

Test Predection


```python
import numpy as np

sample= np.array([[2022, 5, 12, 1, 0]])
prediction = model.predict(sample)

print("Prediction", prediction)
```
